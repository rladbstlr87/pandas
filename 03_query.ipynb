## where()
- 조건이 False인 데이터들을 일괄수정하는 함수

```python
cond = df['fare'] < 30
df['fare'].where(cond, 1).tail() # 30불보다 작은 사람의 fare를 일괄적으로 1로 바꾼다
```

## isin

```python
cond = df['embarked'].isin(['Q', 'S'])
df.loc[cond].head()
```

### 연습

```python
tips = sns.load_dataset('tips')
tips.head(10)
```

- 남성 흡연자 조건 필터링
- tip을 가장 많이 낸 순서로 내림차순 정렬
- 상위 10개 출력

```python
cond1 = tips['sex'] == 'Male'
cond2 = tips['smoker'] == 'Yes'
tips.loc[cond1 & cond2].sort_values('tip', ascending=False).head(10)
```

- 식사인원 5명 이상인 저녁테이블
- 흡연자가 없는 주말 테이블
- 그 사람들의 total_bil, tip 출력

```python
cond1 = tips['size'] >= 5
cond2 = tips['time'] == 'Dinner'
cond3 = tips['smoker'] == 'No'
cond4 = tips['day'].isin(['Sat', 'Sun'])

tips.loc[(cond1 & cond2) & (cond3 & cond4), ['total_bill', 'tip']].head()
```
