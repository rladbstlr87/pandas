```python
import pandas as pd
import seaborn as sns
```

```python
df = sns.load_dataset('penguins')
```

```python
df.head()
```

```python
# 5000g 이상의 펭귄 출력
df_copy = df.copy()
cond1 = df_copy['body_mass_g'] >= 5000
df_copy[cond1].sort_values('bill_length_mm', ascending=False)
```

```python
# Gentoo종의 flipper_length 평균
df_copy = df.copy()
cond2 = df_copy['species'] == 'Gentoo'
df_copy.loc[cond2, 'flipper_length_mm'].mean()
```
