```python
df_copy.head() # 운동량은 오브젝트로 되어있는데 이걸 숫자로 바꿔주려면...
```

```python
pd.to_numeric(df_copy['운동량'])
```

```python
df_copy.loc[2344] # 운동량 탄소량을 보면 /N이라고 입력되있어서 이걸 전처리 해주어야 to numeric이 작동한다
```

```python
df_copy['운동량'] = pd.to_numeric(df_copy['운동량'], errors='coerce')
df_copy.info() # 운동량의 dtype이 object에서 float로 변경되엇따
```

## cut()
지정한 구분선대로 데이터를 나누는 것

```python
bins = [0, 100, 200, 300, df_copy['운동량'].max()] # 리스트의 범위를 먼저 만들어봤다
pd.cut(df_copy['운동량'], bins) # 163936052.3 은 이상치여서 보통 제거하고 다시 계산을 한다
```

```python
pd.cut(df_copy['운동량'], bins=10)
pd.cut(df_copy['운동량'], bins=10).value_counts() # 총 30만건의 데이터를 10% 분위별 구간을 보고싶은건데 이상치때문에 좋지 못하다
```

```python
pd.qcut(df_copy['운동량'], q=10).value_counts() # 32만개를 10개로 나눴으므로 3만2천개씩 구간을 보여준다
```

```python
# qcut을 쓰면 우리가 보는 표준분포처럼 나타낼 수도 있다
qcut_bins = [0, 0.2, 0.8, 1]
pd.qcut(df_copy['운동량'], qcut_bins).value_counts()
```

```python
df_copy['운동량_한글'] = pd.qcut(df_copy['운동량'], qcut_bins, labels=['적음', '보통', '많음'])
df_copy.head(2)
```
