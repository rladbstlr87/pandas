```python
import seaborn as sns
import pandas as pd

df = sns.load_dataset('titanic')
df_copy = df.copy()
```

```python
df.head(3)
```

# groupby()
- 성별로 묶어서 나머지 컬럼들의 평균을 보는거다 생존의 평균이나 페어의 평균이나./.

```python
# 남자 생존자의 각 컬럼별 평균, 여자 생존자의 각 컬럼별 평균
df_copy.groupby('sex').mean(numeric_only=True)
```

```python
df_copy.groupby(['sex', 'pclass']).mean(numeric_only=True)
# 앞에서부터 묶고싶은 컬럼을 정하고 나머지 순차적으로.. 큰 분류부터 작은 분류 순으로
```

```python
# 다른 컬럼은 제외하고 보고싶으면 1
df_copy.groupby(['sex', 'pclass'])['survived'].mean(numeric_only=True)
```

```python
# 다른 컬럼은 제외하고 보고싶으면 2
df_copy.groupby(['sex', 'pclass'])[['survived']].mean(numeric_only=True)
```

## reset_index()
- 각 값에 접근하려면 그루핑되어있어서 풀어줘야 한다

```python
df_temp = df_copy.groupby(['sex', 'pclass'])[['survived']].mean()
df_temp.reset_index()
```

```python
df_copy.groupby(['sex', 'pclass'])[['survived', 'age']].agg(['mean', 'sum'])
# agg로 보고싶은 결과를 여러개 제시함으로써 groupby를 한 번더 groupby 햇다고 보면 된다
```
