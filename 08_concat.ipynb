## concat()
- dataframe 끼리 직렬 연결할 때

```python
pd.concat([df1_copy, df2_copy])
```

- 보통 인덱스번호는 따로쓰지않고 통으로 묶을텐데 그럴땐

```python
pd.concat([df1_copy, df2_copy], ignore_index=True).tail(2) # ignore_index를 True로 설정하면 된다
```

```python
sales1 = df1_copy.iloc[:, :5] # 열을 가운데 기준으로 나눠서 오른쪽을 선택한 것
sales2 = df2_copy.iloc[:, 5:] # 열을 가운데 기준으로 나눠서 왼쪽을 선택한 것
```

```python
sales1.info() # 컬럼이 어떻게 나눠졌는지를 유의하여 보자
```

```python
sales2.info() # 이제 이 두개를 병렬로 연결해볼거다
```

## merge()

```python
info = {
    '고객명': ['박세리', '이대호', '손흥민', '김연아', '마이클조던'],
    '생년월일': ['1980-01-02', '1982-02-22', '1993-06-12', '1988-10-16', '1970-03-03'],
    '성별': ['여자', '남자', '남자', '여자', '남자']
}

money = {
    '고객명': ['김연아', '박세리', '손흥민', '이대호', '타이거우즈'],
    '연봉': ['2000원', '3000원', '1500원', '2500원', '3500원']
}
```

```python
df1 = pd.DataFrame(info)
df2 = pd.DataFrame(money)
df1
```

```python
df2
```

```python
info_money = pd.merge(df1, df2)
info_money
```

```python
info = {
    '이름': ['박세리', '이대호', '손흥민', '김연아', '마이클조던'],
    '생년월일': ['1980-01-02', '1982-02-22', '1993-06-12', '1988-10-16', '1970-03-03'],
    '성별': ['여자', '남자', '남자', '여자', '남자']
}

money = {
    '고객명': ['김연아', '박세리', '손흥민', '이대호', '타이거우즈'],
    '연봉': ['2000원', '3000원', '1500원', '2500원', '3500원']
}
# 두개의 이름, 고객명의 key값은 다른데 value 속성은 같을때의 merge를 보자
df1 = pd.DataFrame(info)
df2 = pd.DataFrame(money)
```

```python
pd.merge(df1, df2, left_on='이름', right_on='고객명', how='outer')
```
